```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import RK45,solve_ivp
import scipy.constants as const
```

```python
tmax= 100  #in years
n=100000  #no. of points

times=np.linspace(0,tmax,n)
```

```python
a = 0.4   # Reproduction rate of prey
b = 0.001   # Predator feeding rate per prey creature
d = 0.003  # Reproduction rate of predators per prey species
c = 0.6   # Predator mortality rate when prey is not available

r_=c/d
f_=a/b

r0= 600
f0= 500

R0 = r_ + 10
F0 = f_ + 10

R_1=0
F_1=0




print(r_,f_)
```

```python
def lotkavoltera(t,y):
    r,f = y
    drdt = a*r - b*r*f
    dfdt = -c*f + d*r*f
    return[drdt,dfdt]
```

```python
sol = solve_ivp(lotkavoltera,t_span=[0,tmax],y0=[r0,f0],t_eval=times,method="Radau" ,dense_output=True)
sol1 = solve_ivp(lotkavoltera,t_span=[0,tmax],y0=[R0,F0],t_eval=times ,method="Radau",dense_output=True)
sol2 = solve_ivp(lotkavoltera,t_span=[0,tmax],y0=[R_1,F_1],t_eval=times,method="Radau" ,dense_output=True)
sol3= solve_ivp(lotkavoltera,t_span=[0,tmax],y0=[r_,f_],t_eval=times,method="Radau" ,dense_output=True)
    
t=sol.t
r=sol.y[0]
f=sol.y[1]
plt.plot(t,r,"pink",label="Rabbits")
plt.plot(t,f,"orange",label="Foxes")
plt.xlabel("Time")
plt.title(r"$\mathbf{Population\ vs\ Time}$")
plt.ylabel("Population of species")

plt.legend(loc=1)

plt.savefig('lvmodelplot1.png', dpi=200)

plt.show()

T=sol1.t
R=sol1.y[0]
F=sol1.y[1]
plt.plot(T,R,"pink",label="Rabbits")
plt.plot(T,F,"orange",label="Foxes")
plt.xlabel("Time")
plt.title(r"$\mathbf{Population\ vs\ Time}$")
plt.ylabel("Population of species")

plt.legend(loc=1)
plt.savefig('lvmodelplot2.png', dpi=200)
plt.show()

T2=sol2.t
R2=sol2.y[0]
F2=sol2.y[1]
plt.plot(T2,R2,"pink",label="Rabbits")
plt.plot(T2,F2,"orange",label="Foxes")
plt.xlabel("Time")
plt.title(r"$\mathbf{Population\ vs\ Time}$")
plt.ylabel("population of species")

plt.legend()
plt.savefig('lvmodelplot3.png', dpi=200)
plt.show()


T3=sol3.t
R3=sol3.y[0]
F3=sol3.y[1]
plt.plot(T3,R3,"pink",label="Rabbits")
plt.plot(T3,F3,"orange",label="Foxes")
plt.xlabel("Time")
plt.title(r"$\mathbf{Population\ vs\ Time}$")
plt.ylabel("population of species")

plt.legend()
plt.savefig('lvmodelplot4.png', dpi=200)
plt.show()


```
